## The first plot: Turbulent Kinetic Energy and Reynolds stresses.

For starters, we will define our fundamental values of viscosity and density of the fluid, which will be standard:

<font size="3">$ρ=1\frac{kg}{m^3}$</font>

<font size="3">$μ=0.00125\frac{kg}{ms}$</font>

on paraview they were defined and calculated the mean values of the velocities which will be very much needed throughout the assignment. The first moment where they are needed are in the Reynolds stresses, where the Reynolds stress shall be defined as:

<font size="3">$τ^R=-ρ(\overline{u'u'})$</font>

The value of the average of u'u' can as well be defined as:

<font size="3">$\overline{u'u'}=\overline{uu}-\overline{u}\overline{u}$
    
The calculation of the values of the Reynolds stresses will be done for $R_u$$_u$, $R_u$$_v$, $R_v$$_v$, $R_u$$_w$, $R_v$$_w$ and $R_w$$_w$, using the respective velocities where needed.
    
For last, in this first paragraph, we will calculate the Turbulent Kinetic Energy, defined as TKE with the following equation:
    
<font size="3">$TKE$=$\frac{1}{2}$($R$$_u$$_u$+$R$$_v$$_v$+$R$$_w$$_w$)</font>
    
We can see below all the graphics of the different values. We can clearly see that the highest peak comes from R$_u$$_u$, which makes sense, as it is the direction where the flow is moving. That is followed by the TKE, and so on.

It's curious how 2 values are null: R$_u$$_w$ and R$_v$$_w$, while R$_u$$_v$ is actually negative.
    
The second plot shows the same studied values, but with a different x-coordinate which is y$^+$ in a logarithmic scale.

```python
plt.figure(figsize=(10,8))
fig = plt.plot(data1['Points:1'],data1['RUU'],color="blue",label='$R_u$$_u$')
fig = plt.plot(data1['Points:1'],data1['RUV'],color="green",label='$R_u$$_v$')
fig = plt.plot(data1['Points:1'],data1['RVV'],color="orange",label='$R_v$$_v$')
fig = plt.plot(data1['Points:1'],data1['RUW'],color="purple",label='$R_u$$_w$')
fig = plt.plot(data1['Points:1'],data1['RVW'],color="brown",label='$R_v$$_w$')
fig = plt.plot(data1['Points:1'],data1['RWW'],color="black",label='$R_w$$_w$')
fig = plt.plot(data1['Points:1'],data1['TKE'],color="red",label='TKE')

plt.grid()
plt.legend(loc=0,fontsize=12)

plt.xlabel('$Distance\ normal\ to\ the\ wall$',fontsize='15')
plt.ylabel('$Rᵢⱼ\ or\ TKE$',fontsize='15')

plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))
plt.xscale('log')
fig = plt.plot(data1['y_plus'],data1['RUU'],color="blue",label='$R_u$$_u$')
fig = plt.plot(data1['y_plus'],data1['RUV'],color="green",label='$R_u$$_v$')
fig = plt.plot(data1['y_plus'],data1['RVV'],color="orange",label='$R_v$$_v$')
fig = plt.plot(data1['y_plus'],data1['RUW'],color="purple",label='$R_u$$_w$')
fig = plt.plot(data1['y_plus'],data1['RVW'],color="brown",label='$R_v$$_w$')
fig = plt.plot(data1['y_plus'],data1['RWW'],color="black",label='$R_w$$_w$')
fig = plt.plot(data1['y_plus'],data1['TKE'],color="red",label='TKE')

plt.grid()
plt.legend(loc=0,fontsize=12)

plt.xlabel('y⁺',fontsize='15')
plt.ylabel('$Rᵢⱼ\ or\ TKE$',fontsize='15')

plt.tight_layout()
```

## Second plot: Laminar, turbulent and total shear stress

This following plot studies 3 types of stress that are: Laminar stress, turbulent stress and total stress. Each of those have an equation which we are defining just below:

<font size="3">$τ_{lam}$=$μ\frac{du}{dy}$</font>
    
<font size="3">${τ}_{turb}$$=-$$R_{uv}$</font>

<font size="3">${τ}_{total}={τ}_{lam}+{τ}_{turb}$</font>

As the plot will be shown below, the x-coordinate is y+ with a normal scaling, no logarithmic scaling this time.

```python
plt.figure(figsize=(14,8))
fig = plt.plot(data2['y_plus'],data2['shear_lam'],color="red",label='laminar stresses')
fig = plt.plot(data2['y_plus'],data2['shear_turb'],color="blue",label='turbulent stresses')
fig = plt.plot(data2['y_plus'],data2['shear_tot'],color="green",label='total stresses')

plt.grid()
plt.legend(loc=0,fontsize=12)

plt.xlabel('y⁺',fontsize='15')
plt.ylabel('Stresses',fontsize='15')

axes = plt.gca()
axes.set_xlim([0,950])

plt.tight_layout()
```

## Third plot: $u^+$

Here we shall define 2 values, which are $y^+$ and $u^+$ which will be showcased by defining the following equations which are required to find those values to plot them, and $y^+$ in particular for the ability to plot this and the previous plot we discussed:

First and foremost, $u_τ$ has to be found manually within Paraview, searching for the wall shear stress at the point where we are studying (on the x-axis we are studying at the point of 900), so it's needed to find the laminar shear stress where y=0. Afterwards, we will need to calculate the $u_τ$ with the following equation:


<font size="4">$u_τ$=$\sqrt\frac{τ_w}{ρ}$</font>

<font size="4">$u^+$=$\frac{\overline{u}}{u_τ}$</font>

<font size="4">$y^+$=$\frac{{yu_τ}}{ν}$</font>

And so we can define, this time with x-axis having $y^+$ in logarithmic scales, the following graph that shows the increase of the $u^+$ at the growth of the value at the horizontal axis.

```python
plt.figure(figsize=(14,8))
fig = plt.plot(data3['y_plus'],data3['u_plus'],color="orange")

plt.xscale('log')
plt.xlabel('y⁺',fontsize='15')
plt.ylabel('u⁺',fontsize='15')

plt.grid()

axes = plt.gca()
axes.set_xlim([0.1,1000])

plt.tight_layout()
```
